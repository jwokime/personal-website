```python
import altair as alt
from vega_datasets import data
import pandas as pd
```

```python
df = pd.read_csv("countries-visited.csv")
df
```

```python
# Data generators for the background
sphere = alt.sphere()
graticule = alt.graticule()

# Source of land data
source = alt.topo_feature(data.world_110m.url, 'countries')

# Layering and configuring the components
alt.renderers.set_embed_options(actions=False)
chart = alt.layer(
    alt.Chart(sphere).mark_geoshape(fill=''),
    alt.Chart(graticule).mark_geoshape(stroke='gray', strokeWidth=0.5),
    alt.Chart(source).mark_geoshape(fill='black', stroke='gray').encode(
        tooltip=['Country:N', 'Comments:N']
    ).transform_lookup(
        lookup='id',
        from_=alt.LookupData(data=df, key='id', fields=['Country', 'Comments']),
        default=None
    )
).project(
    'naturalEarth1'
).properties(width=360, height=215)

chart
```

```python
chart.save('worldmap.html')
```
